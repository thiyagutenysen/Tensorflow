```python
# making changes to the data
```

```python
# create a new column named power
# method1
data["Power"] = data["HP"]+data["Attack"]+data["Defense"]+data["Sp. Atk"]+data["Sp. Def"]+data["Speed"]
data.head()
```

```python
data = data.drop(columns=["Power"]) # helper code

# create a new column named power
# method2
data["Power"] = data.iloc[:, 4:10].sum(axis=1) # axis=1, bcz we need to add horizontally
data.head()
```

```python
# delete a column
# we can specify any number of columns by using comma "," inside the column list
# we need to assign the changes to the dataframe otherwise nothing will change
data = data.drop(columns=["Power"])
data.head()
```

```python
# helper code
data["Power"] = data.iloc[:, 4:10].sum(axis=1) # axis=1, bcz we need to add horizontally
```

```python
# reorder columns
# put power from last to after speed
print("before", data.head())
cols = list(data.columns.values)
data = data[cols[:10]+[cols[-1]]+cols[10:-1]]
print("after", data.head())
```

```python
# Edit the dataframe with conditions
## conditional changes
```

```python
# example 1 
# change fire type pokemon to flamer type pokemon
data.loc[data['Type 1'] == 'Fire', 'Type 1'] = "Flame"
data.head()
```

```python
# example 2
# change grass pokemon with hp more than 50 to be legendary
data.loc[(data['Type 1'] == 'Grass') & (data["HP"]>50), 'Legendary'] = True
data.head()
```

```python
# example 2
# 1 condition, change 2 columns
data.loc[(data['Sp. Atk'] > 100), ['Legendary', 'Generation']] = [True, 5]
data.head()
```
